# Download dataset from HuggingFace
Dataset link - https://huggingface.co/datasets/DebasishDhal99/German_Names_Central_And_Eastern_Europe

- A dataset is made up of single or multiple components (called subsets). In addition it might have train, test, dev, validation splits. The by-default split is train.

- This is done for the sake of modularity. If I want a specific section of the dataset, I can do that, without needing to download the entire dataset (which might be of multiple GBs).

- If a dataset has multiple subsets, you should mention which subset you want to download. In addition, you have the option to mention which split (train/test) you want to download. Regardless, it'll download all the splits of a chosen subset and then generate the split.

```python
df_raw = datasets.load_dataset("DebasishDhal99/German_Names_Central_And_Eastern_Europe", #dataset name
                               "czech_german", #Mention which subset you want to download
                               split = "train" #Optional. But not providing this will convert the variable in which you're storing this, into a dictionary.
                               )
df = df_raw.to_pandas() #Now df is a regular pandas DataFrame, where all pandas operation are applicable.

df.sample(5)
```

# Download a Gated Dataset
Link - https://huggingface.co/datasets/oscar-corpus/OSCAR-2301 (A Gated Dataset)


- To access a Gated Dataset, you need to have permission of the dataset's author(s).
- Visit the link, they usually ask for your name and email. You need to agree to their terms, and your name/email will be registered automatically. Usually you get instant access, as soon as you agree to their terms.
- Now you have the access, but to access the dataset using Python, you need a Read Token.
- Go to you HuggingFace profile->Setting->Access Tokens->New Token->Generate a new token. Now you have that token. Copy it.

```python
df_raw = datasets.load_dataset("oscar-corpus/OSCAR-2201",
                               language = "or", #This is essentially the subset. If some error arises due to 'language', try replacing it with 'lang'/'subset' or remove it altogether.
                               token = "Your Token Here", #Don't share it with others.
                               split = "train"
                               ).to_pandas()
df_raw.sample(10)
```

# Downloading a model from HuggingFace

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "gpt2" #Put a model name of your choice (as it appears in HuggingFace)
tokenizer_name = model_name

tokenizer = AutoTokenizer.from_pretrained(tokenizer_name) #LLM don't understand words, so we convert words into a numerical representation using tokenizer of that specific model.
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
prompt = "I'm trying to make a demo on how to use HuggingFace"

tokens = tokenizer(prompt, return_tensors="pt")
output = model.generate(tokens.input_ids,
                        max_length=300,
                        do_sample=True
                        )
generated_text = tokenizer.decode(output[0],
                                  skip_special_tokens=True
                                  )

print(f"Generated Text:\n{generated_text}")
```

# How to specify subsets in your own dataset?

- Subsets are configured using YAML Syntax, which is indentation sensitive.
- To do that, go to the README file of your dataset, put triple-dashes in two consecutive lines. Now, everything inside these dashes will be your metadata, and must follow YAML Syntax.
- A sample subset configuration looks like this

```python
configs:
  - config_name: polish_german
    data_files:
      - split: train
        path: german_polish.csv
  - config_name: czech_german
    data_files:
      - split: train
        path: german_czech.csv
  - config_name: lithuanian_german
    data_files:
      - split: train
        path: german_lithuanian.csv
  - config_name: latvian_german
    data_files:
      - split: train
        path: german_latvian.csv
  - config_name: estonian_german
    data_files:
      - split: train
        path: german_estonian.csv
  - config_name: slovak_german
    data_files:
      - split: train
        path: german_slovak.csv
  - config_name: slovene_german
    data_files:
      - split: train
        path: german_slovene.csv
```

- Follow this link to see an actual metadata. https://huggingface.co/datasets/DebasishDhal99/German_Names_Central_And_Eastern_Europe/blob/main/README.md

- If you're uploading multiple files, then they should either have identical column names or they should belong to separate subsets. If you don't mention subsets after uploading multiple files, HuggingFace will try to merge them all into one big dataset, and thus, those files need to have the same column names.

# Space in HuggingFace

- HuggingFace offers a bunch of frameworks to build Space. They're usually faster to build as compared to the traditional JS, HTML, CSS, JS Framework route.
- I've used Streamlit and Gradio. They are quite similar, but you can test only Gradio within your local notebook. I prefer Gradio.
- In both of them, you have a main script named *app.py*. At the end of this script, you launch the Space using .launch() method.
- Both Gradio and Streamlit offer you the option of Text, Image, Video, Audio input(s) and output(s). Basically, you don't have to code every single feature yourself, many are already taken care of for you.
- For free-tier spaces, the space sleeps (becomes inactive) after two days of inactivity.
